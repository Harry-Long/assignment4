```python
%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
```

## 1. A line plot showing the epoch accuracy for each of the runs

```python
fig, ax = plt.subplots()
total_time = []
for p in [1,2,4,8]:
    df = pd.read_csv(f"metrics{p}.csv", index_col=0)
    ax.plot(df.epoch_accs, label = f"PPN {p}")
    total_time.append(df.elapsed_time.values[-1])
    
ax.set_xlabel('Epoch')
ax.set_ylabel('Training Accuracy')
_ = plt.legend()
```

## 2. A second line plot that shows the total runtime by the number of processes used

```python
plt.figure()
plt.plot([1,2,4,8], total_time, "bo-")
plt.xlabel("Number of Processes")
plt.ylabel("Total Runtime (secs)")
plt.show()
```

- It's obvious that the training time decreases as we increase the number of processes available.
- This lines up with the expectations of the scalability laws that we discussed, and the obtained speedup is not super linear.
- I think Gustafson-Barsis' Law is most applicable in this case according to the lecture slides, but the speedup is not super-linear as expected due to some reasons. This speedup line looks more consistent with Amdahl's law.
- Our model accuracy in same number of epochs is declined by the increase of the number of processes running.
